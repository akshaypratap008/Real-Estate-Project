## Recommender System - based on TopFacilities

```python
df = data.copy()
```

```python
df.drop(columns = ['NearbyLocations', 'Link'], inplace = True)      # drop unnecessary columns
```

```python
df[['PropertyName', 'TopFacilities']]
```

```python
df['TopFacilities'].iloc[0]
```

```python
import re
def extract_facilities(row):
    return re.findall("'(.*?)'", row)       # extract everything inside single inverted comas and return list

df['TopFacilities'] = df['TopFacilities'].apply(extract_facilities)
```

```python
df.head()
```

```python
# convert the list into strings

df['TopFacilities'] = df['TopFacilities'].apply(' '.join)
```

```python
df['TopFacilities'].iloc[0]
```

applying tfidf vectoriser to TopFacilities column

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(stop_words='english', ngram_range=(1,2))
tfidf_matrix = tfidf.fit_transform(df['TopFacilities'])

```

```python
tfidf_matrix.toarray()
```

```python
# cosine similarity of each vector with allthe other vectors 

from sklearn.metrics.pairwise import cosine_similarity

cos_sim1 = cosine_similarity(tfidf_matrix, tfidf_matrix)
```

```python
def recommend_properties(property_name, cosine_sim = cos_sim1):
    # get the index of the property name
    idx = df.index[df['PropertyName'] == property_name].tolist()[0]
    
    # similarity score of the idx item with respect to other items. Using enumerate to also pull the position of the elements
    sim_scores = list(enumerate(cos_sim1[idx]))
    
    

    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse= True)       # sorting in descending order based on similarity score which is at index 1 in each tuple inside the big list 

    del sim_scores[0]       # remove the 0th index as it is the similarity score with itself
    
    # extract top 5 simarity scores and their indicies
    sim_scores = sim_scores[:5]

    # extract the indicies from the list of tupples sim_scores
    sim_property_indicies = [i[0] for i in sim_scores]

    recomendations_df = pd.DataFrame({
        'PropertyName': df['PropertyName'].iloc[sim_property_indicies],
        'SimilarityScore': sim_property_indicies
    })

    return recomendations_df

    
```

```python
def add_data_refined_df(data, row_index):
    for i in data.keys():       # navigate to each key in outer dict (eg 2 BHK and 3 BHK in data2 variable)
        for j in data[i]:           # inside inner dict, navigate to every inner keys( eg 'building_type', 'area_type', etc)
            col_name = f'{j}_{i}'       # create a column name (eg building_type_2 BHK, area_type_3 BHK)
            if col_name not in refined_df.columns:      # check if the column already exist
                refined_df[col_name] = None                # if the column doesn't exist, create a column and give it None value  
            refined_df.loc[row_index, col_name] = data[i][j]      # update the value of that column

    return refined_df
```

```python
def range_split_area(text):
    if text is not None:
        text_list = str(text).strip().split('-')
        if len(text_list) == 1:
            upper_area = float(text_list[0].strip().replace(',', '').split(' ')[0])
            lower_area = float(text_list[0].strip().replace(',', '').split(' ')[0])
        else:
            lower_area = float(text_list[0].strip().replace(',', '').split(' ')[0])
            upper_area = float(text_list[1].strip().replace(',', '').split(' ')[0])
        return lower_area, upper_area
    else:
        return None
    
```

```python
def split_price(text):
    text_list = str(text).strip().split('-')
    if len(text_list) > 1:
        lower_price_text = text_list[0].strip()
        upper_price_text = text_list[1].strip()
        if 'L' in lower_price_text:
            lower_price = float(re.findall(r"\d+(?:\.\d+)?", lower_price_text)[0]) * 0.01
            upper_price = float(re.findall(r"\d+(?:\.\d+)?", upper_price_text)[0]) * 0.01
        else:
            lower_price = float(re.findall(r"\d+(?:\.\d+)?", lower_price_text)[0])
            upper_price = float(re.findall(r"\d+(?:\.\d+)?", upper_price_text)[0])
        return round(lower_price, 2), round(upper_price,2)

    elif len(text_list) == 1:
        if text_list[0] == 'Price on Request':
            lower_price = 'Price on Request'
            upper_price = 'Price on Request'
            return lower_price, upper_price
        elif (text_list[0] is not None) or (text_list[0] != 'nan'):
            price_text = text_list[0].strip()
            if 'L' in price_text:
                price = float(re.findall(r"\d+(?:\.\d+)?", price_text)[0])
                lower_price = price
                upper_price = price
                return round(lower_price,2), round(upper_price, 2) 
                       
        
```

```python
def recommend_properties2(property_name, top_n=5):
    
    # Get the similarity scores for the property using its name as the index
    sim_scores = list(enumerate(cos_sim2[ohe_df_scaled.index.get_loc(property_name)]))
    
    # Sort properties based on the similarity scores
    sorted_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    
    # Get the indices and scores of the top_n most similar properties
    top_indices = [i[0] for i in sorted_scores[1:top_n+1]]
    top_scores = [i[1] for i in sorted_scores[1:top_n+1]]
    
    # Retrieve the names of the top properties using the indices
    top_properties = ohe_df_scaled.index[top_indices].tolist()
    
    # Create a dataframe with the results
    recommendations_df = pd.DataFrame({
        'PropertyName': top_properties,
        'SimilarityScore': top_scores
    })
    
    return recommendations_df
    
```
