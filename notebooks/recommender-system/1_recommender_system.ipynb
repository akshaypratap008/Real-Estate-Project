## Recommender System - based on TopFacilities

```python
df = data.copy()
```

```python
df.drop(columns = ['NearbyLocations', 'Link'], inplace = True)      # drop unnecessary columns
```

```python
df[['PropertyName', 'TopFacilities']]
```

```python
df['TopFacilities'].iloc[0]
```

```python
import re
def extract_facilities(row):
    return re.findall("'(.*?)'", row)       # extract everything inside single inverted comas and return list

df['TopFacilities'] = df['TopFacilities'].apply(extract_facilities)
```

```python
df.head()
```

```python
# convert the list into strings

df['TopFacilities'] = df['TopFacilities'].apply(' '.join)
```

```python
df['TopFacilities'].iloc[0]
```

applying tfidf vectoriser to TopFacilities column

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(stop_words='english', ngram_range=(1,2))
tfidf_matrix = tfidf.fit_transform(df['TopFacilities'])

```

```python
tfidf_matrix.toarray()
```

```python
# cosine similarity of each vector with allthe other vectors 

from sklearn.metrics.pairwise import cosine_similarity

cos_sim1 = cosine_similarity(tfidf_matrix, tfidf_matrix)
```

```python
def recommend_properties(property_name, cosine_sim = cos_sim1):
    # get the index of the property name
    idx = df.index[df['PropertyName'] == property_name].tolist()[0]
    
    # similarity score of the idx item with respect to other items. Using enumerate to also pull the position of the elements
    sim_scores = list(enumerate(cos_sim1[idx]))
    
    

    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse= True)       # sorting in descending order based on similarity score which is at index 1 in each tuple inside the big list 
    print(sim_scores)
    del sim_scores[0]       # remove the 0th index as it is the similarity score with itself
    print(sim_scores)
    
    # extract top 5 simarity scores and their indicies
    sim_scores = sim_scores[:5]

    # extract the indicies from the list of tupples sim_scores
    sim_property_indicies = [i[0] for i in sim_scores]

    recomendations_df = pd.DataFrame({
        'PropertyName': df['PropertyName'].iloc[sim_property_indicies],
        'SimilarityScore': sim_property_indicies
    })

    return recomendations_df

    
```

```python
print(recommend_properties("DLF The Arbour"))
```
