```python
df.drop(columns = ['store room', 'floor_category', 'balcony'], inplace = True)
```

```python
df['agePossession'].value_counts()
```

```python
df['agePossession'].replace({
    'relatively new': 'new',
    'relatively old': 'old'
}, inplace = True)
```

```python
df['agePossession'].value_counts()
```

```python
df['furnishing_type'].value_counts()
```

```python
df['furnishing_type'].replace({
    'unfurnished': 0,
    'semifurnished': 1,
    'furnished': 2
}, inplace = True)
```

```python
df.info()
```

```python
df.head()
```

```python
df['property_type'] = df['property_type'].replace({
    'flat': 0,
    'house': 1
})
```

```python
df.head()
```

```python
df['luxury_category'].value_counts()
```

```python
df['luxury_category'] = df['luxury_category'].replace({
    'budget': 0,
    'semi-luxury': 1, 
    'luxury': 2
})
```

```python
df.head()
```

```python
# ohe on sector and agePossesion

df_transformed = pd.get_dummies(df, columns = ['sector', 'agePossession'], drop_first=True, dtype=int)
```

```python
# split X and y

X = df_transformed.drop(columns = 'price')
y = df_transformed['price']
```

```python
y_log = np.log1p(y)
```

```python
y_log
```

```python
# scaling the data
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_scaled = pd.DataFrame(X_scaled, columns = X.columns)
```

```python
# train Linear Regression model

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import KFold, cross_val_score

kfold = KFold(n_splits=10, shuffle=True, random_state=42)
scores = cross_val_score(LinearRegression(), X_scaled, y_log, cv = kfold, scoring='r2')
```

```python
scores.mean()
```

```python
scores.std()
```

```python
lr = LinearRegression()
lr.fit(X_scaled, y_log)
```

```python
coef_df = pd.DataFrame(lr.coef_.reshape(1, 103), columns = X.columns).stack().reset_index().drop(columns = 'level_0').rename(columns = {'level_1': 'feature_name', 0:'coef'})
```
