```python
# drop useless columns like link and property_id
df.drop(columns = ['link', 'property_id'], inplace=True)
```

```python
# rename area column to price_per_sqft
df.rename(columns={'area': 'price_per_sqft'}, inplace = True)
```

### Society

```python
df['society'].nunique()
```

```python
df['society'].sort_values()
```

many society names contain rating in the end

```python
#removing rating from society name

import re
df['society'] = df['society'].apply(lambda name : re.sub(r'\d+(\.\d+)?\s?★', '', str(name))).str.strip().str.lower()
```

```python
df['society'].nunique()
```

### Price columns

```python
df['price'].value_counts().sort_index()
```

Price on Request    11 --> remove these rows

```python
temp_df = df[df['price'] == 'Price on Request']
```

```python
temp_df.index
```

```python
# dropping the rows where price is not available
df = df.drop(index = temp_df.index)
```

```python
df.shape
```

remove 'Lac' and 'Crore' convert them to crores

```python
#remove 'Lac'

mask = df['price'].str.contains('Lac', na = False)
df['price'] = df['price'].astype('object')
df.loc[mask, 'price'] = df.loc[mask, 'price'].str.split().str.get(0).astype(float) * 0.01
```

```python
# remove 'Crore'

mask = df['price'].str.contains('Crore', na = False)
df.loc[mask, 'price'] = df.loc[mask, 'price'].str.split().str.get(0)
```

```python
df['price'] = df['price'].astype(float).round(2)
```

```python
df.rename(columns={'price': 'price_in_cr'}, inplace= True)
```

```python
df.head()
```

### price_per_sqft column

```python
# extract numbers from the sqft columns and convert them to 'Int64'. Chose 'Int64' because it works with missing values

df['price_per_sqft'] = df['price_per_sqft'].str.extract(r'([\d,]+)')[0].str.replace(',', '').str.strip().astype('Int64')
```

```python
df['price_per_sqft'].isna().sum()
```

```python
df.head()
```

### AreaWithType column

```python
df['areaWithType'].value_counts()
```

### Bedroom, Bathrooms, Balcony columns

```python
# bedroom column
df['bedRoom'] = df['bedRoom'].str.split(' ').str.get(0).str.strip().astype('Int64')
```

```python
# all the 9 rows with null values in bedroom have rest of the data missing as well. Therefore removing those rows

df = df[~df['bedRoom'].isna()]
df.shape
```

```python
# bathroom column
df['bathroom'] = df['bathroom'].str.split(' ').str.get(0).str.strip().astype('Int64')
```

```python
df['bathroom'].isnull().sum()
```

```python
# balcony column
df['balcony'] = df['balcony'].str.strip().str.split(' ').str.get(0).str.replace('No', '0')
```

```python
df['balcony'].isna().sum()
```

```python
df.head()
```

### AdditionalRoom Column

```python
df['additionalRoom'].value_counts()
```

```python
df['additionalRoom'].isna().sum()
df['additionalRoom'] = df['additionalRoom'].fillna('Not Available').str.lower()
```

```python
df.head()
```

### FloorNum Column

```python
df['floorNum'].value_counts().sort_index()
```

```python
df[df['floorNum'].isna()]
```

```python
# independent_houses.csv only contains the floor number on which the flat is located in noOfFloor column. Therefor, only extracting the floor number on which the flat is located

df['floorNum'] = df['floorNum'].str.strip().str.split('of').str.get(0).str.strip().str.lower().replace('ground', '0').replace('basement', '-1').replace('lower', '0').replace('lower ground', '0').str.extract(r'(\d+)').astype('Int64')
```

```python
df['floorNum'].isna().sum()
```

```python
df.head()
```

### Facing column

```python
df['facing'] = df['facing'].str.lower()
```

```python
df['facing'].isna().sum()
```

```python
df['facing'] = df['facing'].fillna('NA')
```

### Adding area column from price and sqft

```python
df.insert(loc=4, column='area', value= ((df['price_in_cr']*10000000) / df['price_per_sqft']).round(2))
```
