```python
df['society']=df['society'].replace('nul', 'independent')
```

### Price column

```python
df['price'].value_counts().sort_index()
```

```python
# drop price on request
df = df[df['price'] != 'Price on Request']
```

```python
df['price'] = df['price'].str.strip().str.lower()
```

```python
# convert prices to crores and remove 'crore' and 'lac'

def treat_price(price):
    if type(price) == float:
        return price
    else:
        if price[1] == 'lac':
            return float(price[0]) * 0.01
        else:
            return float(price[0])
```

```python
df['price'] = df['price'].str.strip().str.split(' ').apply(treat_price).round(2)
```

```python
df['price'].isna().sum()
```

```python
df.rename(columns = {'price': 'price_in_cr'}, inplace=True)
```

```python
df.head()
```

### price_per_sqft column

```python
df['price_per_sqft'] = df['price_per_sqft'].str.strip().str.extract(r'([\d,]+)')
```

```python
df['price_per_sqft'] = df['price_per_sqft'].str.replace(',', '')
```

### bedroom, bathroom, balcony column

```python
# bedroom
df['bedRoom'] = df['bedRoom'].str.strip().str.replace('Bedrooms', '').str.replace('Bedroom', '').str.strip().astype('Int64')
df = df[~df['bedRoom'].isnull()]
```

```python
df.head()
```

```python
#bathroom

df['bathroom'] = df['bathroom'].str.strip().str.split(' ').str.get(0).astype('Int64')
df['bathroom'].isnull().sum()
```

```python
# balcony
df['balcony'] = df['balcony'].str.replace('No', '0').str.strip().str.split(' ').str.get(0)
```
